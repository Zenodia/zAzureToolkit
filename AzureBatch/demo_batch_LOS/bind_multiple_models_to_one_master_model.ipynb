```python
import os
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import pickle
```

```python
#encironment used wjpy36
import numpy as np
import time
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import random
import pandas as pd
SEED = 2019
random.seed(SEED)
np.random.seed(SEED)

import warnings
warnings.filterwarnings("ignore")

#==============================================================================
# Data 
#==============================================================================
df=pd.read_csv('flat_table.csv')
df.head()
```

```python
cols=list(df.columns)
len(cols),cols
```

```python
#downloaded pickled models
os.listdir('./pkl_models/')
```

```python
n=len(cols)
label=cols[-1]
features=cols[:-1]
X=df.loc[:,features].values
y=df.loc[:,label].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=123456)
X.shape, y.shape, X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# fetch all models in a list
models=[]
for file in os.listdir('./pkl_models/'):
    if file.endswith('.pkl') :
        with open('./pkl_models/'+file, 'rb') as f:  
            clf = pickle.load(f)
            models.append(clf)

print(models, len(models), type(models))
```

```python
i=0
clfs=[]
for  m in models:
    name=str(type(m)).split(' ')[1].split('.')[3]
    if (name[:-2]=='RandomForestClassifier') or (name[:-2]=='DecisionTreeClassifier')  :
        l=len(name)
        print(name,len(name))
        clfs.append((name[:-2]+str(i+1),m))
        m.set_params(min_impurity_decrease=0)
        i+=1
    else:
        l=len(name)
        print(name,len(name))
        print(name)
        clfs.append((name[:-2]+str(i+1),m))
        i+=1
        

    
clfs
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier, VotingClassifier

master_model = VotingClassifier(estimators=clfs, voting='hard')
master_model=master_model.fit(X_train,y_train)
```

```python
y_pred=master_model.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
